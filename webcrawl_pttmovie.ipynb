```python
import requests 
from bs4 import BeautifulSoup
```

```python
url="https://www.ptt.cc/bbs/movie/index.html"
# 使用 GET 方式下載普通網頁
res = requests.get(url)
```

```python
soup = BeautifulSoup(res.text, 'html.parser')
print(soup.prettify())
```

```python
#使用select找
tags = soup.select('div.title a')
for t in tags:
    print(t.text)
```
